```python
import pandas as pd
import os.path
from os import pardir, getcwd

season = pd.read_csv(os.path.dirname(os.getcwd()) + '\\ultiscores\\regular_season_scores_20160922.csv')
season = season[season['div']=='Men']

#remove seeds from team names
season['home'] = season['home'].apply(lambda x: str(x).split('(')[0])
season['away'] = season['away'].apply(lambda x: str(x).split('(')[0])
#remove games from cancelled tournaments => teamname is 'nan'
season = season[season['home'] != 'nan']
season = season[season['away'] != 'nan']
season = season[~season['homescore'].isin(['W','L','F'])]
season = season[~season['awayscore'].isin(['W','L','F'])]
season[['homescore','awayscore']] = season[['homescore','awayscore']].astype(int)
season = season[season['homescore'].add(season['awayscore']) > 0]
season.head
#print sum(season['homescore'].add(season['awayscore']) == None)
```

```python
import networkx as nx
G = nx.MultiDiGraph()

teams = pd.unique(pd.concat([season['home'], season['away']], axis=0))

scorediff = season['homescore'].subtract(season['awayscore'])

winnerleft = season.loc[(scorediff > 0),'home']
winnerright = season.loc[scorediff<0,'away']
winners = pd.concat([winnerleft, winnerright]).values.tolist()

loserleft = season.loc[scorediff>0, 'away']
loserright = season.loc[scorediff<0,'home']
losers = pd.concat([loserleft,loserright]).values.tolist()
#what's up with the one tied game? 0-0?
scorediff = scorediff.loc[scorediff!=0]

nodelist = [(i, {'team':t}) for i,t in enumerate(teams)]
edgelist = [(winner, loser, {'weight':weight}) for winner, loser, weight in zip(winners,losers,scorediff)]

G.add_nodes_from(nodelist)
G.add_edges_from(edgelist)
```

```python
ranking = nx.pagerank_numpy(G, alpha=.9)
print ranking.values()
#team_ranking = {nodelist[i][1]['team']: ranking[value] for i, value in enumerate(ranking.values())}
def newkey(old_dict, old_key, new_key):
    old_dict[new_key] = old_dict.pop(old_key)
    return old_dict

#team_ranking = [newkey(team_ranking, i, nodelist[i][1]['team']) for i in range(len(nodelist))]
#print nodelist[0][1]['team']
#print team_ranking
```
